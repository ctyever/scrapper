```python
def strip_comma(x):    
    x = x.strip().replace(',', '')
    # print(x)
    return x
```

```python
data['거래금액(만원)'] = data['거래금액(만원)'].apply(lambda x: int(strip_comma(x)))
```

```python
data = data[['거래금액(만원)', '단지명', '전용면적(㎡)']]
```

```python
data.head(3)
```

```python
data_group_price = data.groupby(['단지명', '전용면적(㎡)']).min()
```
